## Tensorflow Implementation

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow_probability as tfp
tfd = tfp.distributions
```

```python
tf.reset_default_graph()
```

```python
def encoder(ip_data,latent_size):
    x = tf.layers.flatten(ip_data)
    x = tf.layers.dense(x,164,tf.nn.relu)
    x = tf.layers.dense(x,164,tf.nn.relu)
    mean = tf.layers.dense(x,latent_size)
    std = tf.layers.dense(x,latent_size,tf.nn.softplus) # for many epochs relu gives -inf, no activation works but
                                                          # not as good results  
    return tfd.MultivariateNormalDiag(mean,std)
    
```

```python
def prior(latent_size):
    mu = tf.zeros(latent_size)
    sig = tf.ones(latent_size)
    return tfd.MultivariateNormalDiag(mu,sig)
```

```python
def decoder(latent_data,op_size):
    x = tf.layers.dense(latent_data,164,tf.nn.relu)
    x = tf.layers.dense(x,164,tf.nn.relu)
    logit = tf.layers.dense(x,op_size[0]*op_size[1])
    logit = tf.reshape(logit,[-1]+op_size)
    return tfd.Independent(tfd.Bernoulli(logit),2)
    
    
```

```python
data = tf.placeholder(tf.float32, [None,28,28])
encoder = tf.make_template('encoder1',encoder)
decoder = tf.make_template('decoder1',decoder)

latent_size = 2
priorr = prior(latent_size)
encoder_op = encoder(data,latent_size)
latent_data = encoder_op.sample()
decoder_op = decoder(latent_data,op_size = [28,28])
likelihood = decoder_op.log_prob(data)
divergence = tfd.kl_divergence(encoder_op,priorr)
elbo = tf.reduce_mean(likelihood-divergence)
```

```python
optimize = tf.train.AdamOptimizer(0.001).minimize(-elbo)

samples = decoder(priorr.sample(10), [28, 28]).mean()
```

```python
def plot_codes(ax, codes, labels):
    ax.scatter(codes[:, 0], codes[:, 1], s=2, c=labels, alpha=0.1)
    ax.set_aspect('equal')
    ax.set_xlim(codes.min() - .1, codes.max() + .1)
    ax.set_ylim(codes.min() - .1, codes.max() + .1)
    ax.tick_params(
      axis='both', which='both', left='off', bottom='off',
      labelleft='off', labelbottom='off')


def plot_samples(ax, samples):
    for index, sample in enumerate(samples):
        ax[index].imshow(sample, cmap='gray')
        ax[index].axis('off')
```

```python
mnist = input_data.read_data_sets('MNIST_data/')
```

```python
fig, ax = plt.subplots(nrows=20, ncols=11, figsize=(10, 20))
with tf.train.MonitoredSession() as sess:
    for epoch in range(20):
        feed = {data: mnist.test.images.reshape([-1, 28, 28])}
        test_elbo, test_codes, test_samples = sess.run([elbo, latent_data, samples], feed)
        print('Epoch', epoch, 'elbo', test_elbo)
        ax[epoch, 0].set_ylabel('Epoch {}'.format(epoch))
        plot_codes(ax[epoch, 0], test_codes, mnist.test.labels)
        plot_samples(ax[epoch, 1:], test_samples)
        for _ in range(600):
            feed = {data: mnist.train.next_batch(100)[0].reshape([-1, 28, 28])}
            sess.run(optimize, feed)
plt.savefig('vae-mnist.png', dpi=300, transparent=True, bbox_inches='tight')
```
